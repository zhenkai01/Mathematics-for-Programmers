```python
tensor_3d = torch.stack([tensor_2d, tensor_2d + 10, tensor_2d - 5])  #  3  2D 
print("3D Tensor (Cube):\n", tensor_3d)
print("Shape:", tensor_3d.shape)  # 
```

```python
#  4D 
tensor_4d = torch.stack([tensor_3d, tensor_3d + 100])  #  2  3D 
print("4D Tensor (Vector of Cubes):\n", tensor_4d)
print("Shape:", tensor_4d.shape)  # 
```

```python
#  2D 
tensor = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)

# 
print("Tensor:\n", tensor)
print("Shape:", tensor.shape)  # 
print("Size:", tensor.size())  # 
print("Data Type:", tensor.dtype)  # 
print("Device:", tensor.device)  # 
print("Dimensions:", tensor.dim())  # 
print("Total Elements:", tensor.numel())  # 
print("Requires Grad:", tensor.requires_grad)  # 
print("Is CUDA:", tensor.is_cuda)  #  GPU 
print("Is Contiguous:", tensor.is_contiguous())  # 
```

```python
#  2D 
tensor = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
print(":\n", tensor)
```

```python
# 1. ****
print("\n")
print(":", tensor[0])  # 
print(":", tensor[0, 0])  # 
print(":", tensor[:, 1])  # 
```

```python
# 2. ****
print("\n")
reshaped = tensor.view(3, 2)  #  3x2
print(":\n", reshaped)
flattened = tensor.flatten()  # 
print(":\n", flattened)
```

```python
# 3. ****
print("\n")
tensor_add = tensor + 10  # 
print(" 10:\n", tensor_add)
tensor_mul = tensor * 2  # 
print(" 2:\n", tensor_mul)
tensor_sum = tensor.sum()  # 
print(":", tensor_sum.item())
```

```python
# 4. ****
print("\n")
tensor2 = torch.tensor([[1, 1, 1], [1, 1, 1]], dtype=torch.float32)
print(":\n", tensor2)

tensor_dot = torch.matmul(tensor, tensor2.T)  # 
print(":\n", tensor_dot)
```

```python
# 5. ****
print("\n")
mask = tensor > 3  # 
print(" 3 :\n", mask)
```

```python
filtered_tensor = tensor[tensor > 3]  # 
print(" 3 :\n", filtered_tensor)
```
